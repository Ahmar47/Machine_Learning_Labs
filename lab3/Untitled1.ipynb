```python
import numpy as np
import matplotlib.pyplot as plt
from utils import *
import copy
import math
%matplotlib inline
```

```python
train_x, train_y = load_data_multi()
train_x.shape
```

```python
w1, w2, b = 1,2,3 
x1 = train_x[:,0]
x2 = train_x[:,1]

#pred
learning_rate = 0.1
m  = train_x[:,0].shape[0]
for i in range(50):
    pred = (w1*x1) + (w2*x2) + b

    #ost = ( np.sum((pred  - train_y)**2))/(2*m)
    error = pred - train_y
    b = b - (learning_rate/m)*(np.sum(error))
    w1 = w1 - (learning_rate/m)*(np.sum((error * x1)))
    w2 = w2 - (learning_rate/m)*(np.sum((error * x2)))

b,w1,w2
```

```python
def min_max_scaling(x):
   return (x - x.min()) / (x.max() - x.min())

x1 = min_max_scaling(x1)

x1 = min_max_scaling(x2)
train_y = min_max_scaling(train_y)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Example data
#train_x = np.array([[1, 2], [2, 3], [3, 4], [4, 5]])  # 4 training examples, 2 features
#train_y = np.array([2, 3, 4, 5])  # 4 training examples, 1 target variable

# Visualize the 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Extract x1, x2, and y values
x1_values = train_x[:, 0]
x2_values = train_x[:, 1]
y_values = train_y

# Plot the points
ax.scatter(x1_values, x2_values, y_values, c='r', marker='o')

# Set labels and title
ax.set_xlabel('Feature 1')
ax.set_ylabel('Feature 2')
ax.set_zlabel('Target')
ax.set_title('3D Scatter Plot of Training Data')

# Show plot
plt.show()


x1_grid, x2_grid = np.meshgrid(np.linspace(x1.min(), x1.max(), 20),
                               np.linspace(x2.min(), x2.max(), 20))

# Calculate prediction for each point on the grid
y_pred = w1 * x1 + w2 * x2 + b

# Plot the regression plane
ax.plot_surface(x1_grid, x2_grid, y_pred, alpha=0.5, color='blue')

plt.show()
```
