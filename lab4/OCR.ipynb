### Finding the Probabilities

### Training Data

##### Probabailty of 1s and 0s

```python
# Compute the Probabilities using smorthing Method


img2_one_prob = ((sum(img2_X_train))+1)/((img2_X_train.shape[0])+2) 
img2_zero_prob =  ((sum(1 - img2_X_train))+1)/((img2_X_train.shape[0])+2)
img4_one_prob = ((sum(img4_X_train))+1)/((img4_X_train.shape[0])+2)
img4_zero_prob = (sum(1- img4_X_train)+1)/((img4_X_train.shape[0])+2)

print(img2_one_prob.shape)
print(img2_zero_prob.shape)
print(img4_one_prob.shape)
print(img4_zero_prob.shape)
```

```python
# Your code here
# p(2|A1,A2... A256) = ( p(A1|2) * p(A2|2) *p(A3|2) *p(A4|2) * p(A5|2) ... p(A256|2) ) *  p(2)
# p(2|A1,A2... A250) = ( p(A1|2) * p(A2|2) *p(A3|2) *p(A4|2) * p(A5|2) ... p(A256|2) ) *  p(4)
 
# A1 == 0 also for A1 == 1 for 'class 2'
# A1 == 0 also for A1 == 1 for Class 4'  it means for each attribute we have to calculte the 4 probabilities 
# 4 * 256 = 1024 probablities 


sparse = np.zeros((4,256))

for col in range(256):
    sparse[0,col] = ((sum(img2_X_train[:,col] == 0))+1)/((img2_X_train.shape[0])+2)
    sparse[1,col] = ((sum(img2_X_train[:,col] == 1))+1)/((img2_X_train.shape[0])+2)
    
    sparse[2,col] = ((sum(img4_X_train[:,col] == 0 ))+1)/((img4_X_train.shape[0])+2)
    sparse[3,col] = ((sum(img4_X_train[:,col] == 1))+1)/((img4_X_train.shape[0])+2)
    
    
    
print(sparse)
```

```python
# Your code here
# 

sparse2 = np.zeros((4,256))
for col in range(256):
    sparse2[0,col] = sum(1 - img2_X_train[:,col])/(img2_X_train.shape[0])
    sparse2[1,col] = sum(img2_X_train[:,col])/(img2_X_train.shape[0])
    
    sparse2[2,col] = sum(1 - img4_X_train[:,col])/(img4_X_train.shape[0])
    sparse2[3,col] = sum(img4_X_train[:,col])/(img4_X_train.shape[0])
    
print(sparse2)
```

#### Class Probabilities

```python
class2_prob = (sum(train_y == 2))/(train_y.shape[0])
class4_prob = (sum(train_y == 4))/(train_y.shape[0])

print(f"Closs2 Probablity: {class2_prob}")
print(f"Closs4 Probablity: {class4_prob}")
```

#### Predictions

```python
pred2_X_train = []
predictions = np.zeros((train_x.shape[0],1))


i = 0 
for row in train_x:
    img2_zero = np.prod(img2_zero_prob[row == 0])
    img2_one = np.prod(img2_one_prob[row == 1])
    pred_img2 = img2_zero  * img2_one * class2_prob
    #print(pred_img2)
    
    img4_zero = np.prod(img4_zero_prob[row == 0])
    img4_one = np.prod(img4_one_prob[row == 1])
    pred_img4 = img4_zero * img4_one *  class4_prob
    #print(pred_img4)
    
    
    if pred_img2 > pred_img4 :
        #print(f"Wrong Predition {i}")
        predictions[i] = 2
    elif pred_img2 < pred_img4 :
        #print(f"Incorrect Predictions  {i}")
        predictions[i] = 4
        
    elif pred_img2 == pred_img4:
        predictions[i] = 1
    i += 1
    
```
