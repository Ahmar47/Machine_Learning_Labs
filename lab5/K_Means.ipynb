### Initialization of k Centroids

```python
k = 3   # Select centroids from the give image pixels
centroids = np.zeros((k,3))
for i in range(k):
    
    row_idx = r.randint(0,arr.shape[0])
    
    centroids[i,:] = arr2D[row_idx,:]
    
print(centroids)
```

```python
import heapq  # to store the min distance centroid efficiently 
```

```python
clusters_arr = np.zeros((arr2D.shape[0],3))

for i in range(100):
    selected_centroid_row = np.zeros(arr2D.shape[0])
    
    for row in range(arr2D.shape[0]):  # CALCULATE THE DISTANCE FOR EACH PIXEL,FROM EACH DISTANCE 
        min_dis_centroids = [] 
        for clus in range(k):
            #print(clus)
            #print(np.sum(np.abs(row - centroids[clus,:])))
            dist = np.sum(np.abs(arr2D[row] - centroids[clus,:])) # CALCULATE THE DISTANCE
            # 
            heapq.heappush(min_dis_centroids, (dist, clus) )  #make min heap on
                                                     # basis tuple[0]("Cluster distance") index, also store the cluster name
                                                     # to get min cluster and its name easily


        #heapq.heapify(min_dis_centroids)
        #print(min_dis_centroids)
        #print(min_dis_centroids[0][1])
       
        selected_centroid_row[row] = min_dis_centroids[0][1] # min distance value on the zero index,(tuple)[1] for "cluster NO"
    print('iteration',i)
  
    print(centroids)
    print()
    for c in range(k): # 
        
        group_rows = arr2D[selected_centroid_row == c]
        
        centroids[c,:] = np.mean(group_rows,axis=0) # Update the centroids WITH MEAN OF SAME CLUSTER (i.e c1, c2)
print('Finished: ') 
print('final centroids ',centroids)
```
